```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

url='https://raw.githubusercontent.com/gabonegabone/matmodel_4k/master/KZN.csv'
df=pd.read_csv(url)
years=np.array(df['Years'])
population=np.array(df['Population'])
plt.plot(df['Years'],df['Population'])

plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Population of Kazan')
plt.show()
```

```python
from scipy.integrate import odeint
def dydt(y,t):
    k=0.02
    return k*y
    
solve=odeint(dydt,y0=1000000,t=years)
solve=solve.flatten()
%matplotlib inline

plt.plot(years, population, label = r"Kazan's population")
plt.plot(years, solve, label = r"Maltus' hard model")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def s(y):
    a=400000000
    b=350
    return a-b*y
def dydt2(y,t):
    return s(y)*y

solve2=odeint(dydt2,y0=15000,t=years)
solve2=solve2.flatten()
%matplotlib inline

plt.plot(years, population, label = r"Kazan's population")
plt.plot(years, solve2, label = r"Maltus' soft model")
plt.grid(True)
plt.legend()
plt.show()
```
