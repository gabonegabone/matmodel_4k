## Жесткая модель Мальтуса.
$x'=kx$

```
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
  k=0.42
  return k*x

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label=r"$x'=kx$")
plt.grid(True)
plt.legend()
plt.show()
```

## Мягкая модель Мальтуса.
$x'=k(x)x$

```
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def k(x):
  a=4
  b=18
  return a-x*b

def dydt(x,t):
  return k(x)*x

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label=r"$x'=k(x)x$")
plt.grid(True)
plt.legend()
plt.show()
```

## Жесткая модель рыболовства.
$x'=x-x^2-c$

```
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
  c=0.05
  return x-x*x-c

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"$x'=x - x^2 - c$")
plt.grid(True)
plt.legend()
plt.show()
```

## Жесткая модель рыболовства с критическим значением в максимуме $g(x)x$
$x'=x-g(x)x-c$

```
#Номер 4
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def g(x):
  a=1
  b=2
  return a*x-b

def dydt(x,t):
  c=0.2
  return x-g(x)*x-c

t=np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve,label=r"$x'=x-g(x)*x-c$")
plt.grid(True)
plt.legend()
plt.show()
```
