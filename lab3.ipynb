##Модель Мальтуса населения земли

$x'=kx$

**Решение 1**

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
def dxdt(x,t):
    k=0.6
    return x-x*x-k*x
t=np.linspace(0,10,100)
solve = odeint(dxdt, 1, t=t)
solve = solve.flatten()
%matplotlib inline

plt.plot(t, solve, label = r"$x' = x-x^2-kx$")
plt.grid(True)
plt.legend()
plt.show()
```

**Решение 2 (метод Эйлера)**

```python
import numpy as np
from matplotlib import pyplot as plt

h=0.1
t2=np.arange(0,10+h,h)

def euler(f,x0,t,h):
        x=[]
        x.append(x0)
        for i in range(1,len(t)):
            xn=x[i-1]+h*f(x[i-1],t[i-1])
            x.append(xn)
        return x
    
euler=euler(dxdt, 1, t, h)

%matplotlib inline
plt.plot(t, euler, label = r"$Euler solve$")
plt.plot(t, solve, label = r"$x' = x-x^2-kx$")
plt.grid(True)
plt.legend()
plt.show()
```
